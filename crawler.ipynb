```python
from bs4 import BeautifulSoup
import requests
```

```python
alls = []
for i in range(5):
    print(i)
    r = requests.get('https://www.petsmart.com/dog/food/?pmin=0.01&srule=best-sellers&start='+str(i*36)+'&sz=36')
    content = r.content
    soup = BeautifulSoup(content)
    print (soup.p)
    for d in soup.findAll('div',class_="product-tile"):
        name = d.find('div', class_="product-name")
#         n = name.find('span', class_ = "w_CW")
        price = d.find('div', class_='price-sales tile-autoship')
        print (price.text[1:6])
#         print (name.text)
#         print (price.text)

        all1=[]

        if name is not None:
            a = name.text
            a = a.replace('\n','')
            all1.append(a)
        else:
            all1.append("unknown-product")

        if price is not None:
            #print(price.text)
            all1.append(price.text[1:6])
        else:
            all1.append('0')
        alls.append(all1)      
```

```python
len(alls)
```

```python
import pandas as pd
# results = []

flatten = lambda l: [item for sublist in l for item in sublist]
df = pd.DataFrame(alls,columns=['Product', 'Price'])
df.to_csv('dog_food_1208.csv')
```
